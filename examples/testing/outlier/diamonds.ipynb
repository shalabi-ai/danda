# How to use danda
The first step you need to do is to collect analysis, study the report, then decide if you want to go with the automatic danda improvements

```python
#from pprint import pprint
#import numpy as np
import pandas as pd
# notice the need to danda import to have accessor df
import danda # noqa: F401  # registers the pandas accessor
from danda.report_renderer import ReportRenderer
```

```python
df = pd.read_csv("diamonds.csv") #"https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv")
df
```

```python
report = df.dg.clean().dg.optimize().dg.analyze().dg.report
renderer = ReportRenderer()
print(renderer.render(report))
#print(json.dumps(report, indent=4, default=str))
```

```python
new_df = df.dg.clean().dg.optimize()
report_new = new_df.dg.report
```

```python
renderer = ReportRenderer()
print(renderer.render(report_new))
```

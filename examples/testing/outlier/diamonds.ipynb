# How to use danda
The first step you need to do is to collect analysis, study the report, then decide if you want to go with the automatic danda improvements

```python
#from pprint import pprint
#import numpy as np
import pandas as pd
# notice the need to danda import to have accessor df
import danda # noqa: F401  # registers the pandas accessor
from danda.report_renderer import ReportRenderer
```

```python
df = pd.read_csv("diamonds.csv") #"https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv")
df
```

## Dana could do this automatically for this data set
- Converted the following columns to category: cut, color, clarity
- save 32.2% of memory
- identified potential numeric columns which can be converted to category: cut, color, clarity
- identified a set of outliers: price, depth, carat, table, z, x, y
- identified a skewed distribution: price

```python
report = df.dg.clean().dg.optimize().dg.analyze().dg.report
renderer = ReportRenderer()
print(renderer.render(report))
#print(json.dumps(report, indent=4, default=str))
```

# let us use other outlier method
IQR finds 6.55% of price values as outliers.
Z-score finds only 2.23%.
The thresholds are very different because IQR is based on quartiles, while Z-score is based on the mean and standard deviation.

```python
df.dg.config.analysis.outlier_method = "zscore"
report = df.dg.clean().dg.optimize().dg.analyze().dg.report
renderer = ReportRenderer()
print(renderer.render(report))
```

```python
new_df = df.dg.clean().dg.optimize()
report_new = new_df.dg.report
```

```python
renderer = ReportRenderer()
print(renderer.render(report_new))
```

# How to use danda
The first step you need to do is to collect analysis, study the report, then decide if you want to go with the automatic danda improvements

```python
#from pprint import pprint
#import numpy as np
import pandas as pd
# notice the need to danda import to have accessor df
import danda # noqa: F401  # registers the pandas accessor
from danda.report_renderer import ReportRenderer
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
df
```

```python
report = df.dg.clean().dg.optimize().dg.analyze().dg.report
renderer = ReportRenderer()
print(renderer.render(report))
#print(json.dumps(report, indent=4, default=str))
```

Looking at the report we can see:
* Stripped leading/trailing whitespace from columns: Name
* convert these columns to boolean Survived
* Converted the following columns to category: Pclass, Sex, Age, SibSp, Parch, Embarked
* Memory saved     : 47,066 bytes (38.7%)

Looking at the analysis:
* we have 891 rows and 12 columns
* age has 177missing values
* cabin has 687 missing values

This is good so let us do it



```python
new_df = df.dg.clean().dg.optimize()
```

```python
import numpy as np
import pandas as pd

df = pd.DataFrame({
    # Will become boolean
    "active": [
        "True", "False", "True", "False",
        "True", "False", "True", "False",
        None, None
    ],

    # Will become datetime
    "created_at": [
        "2024-01-01",
        "2024-01-02",
        "2024-01-03",
        "2024-01-04",
        "2024-01-05",
        "2024-01-06",
        "2024-01-07",
        "2024-01-08",
        None,
        None,
    ],

    # Will become numeric
    "price": [
        "10.5",
        "20.1",
        "30.2",
        "40.3",
        "50.4",
        "60.5",
        "70.6",
        "80.7",
        None,
        None,
    ],

    # Will become category (2 unique values / 10 rows = 20%)
    # If your threshold is 10%, use 20 rows instead.
    "country": [
        "Germany",
        "Germany",
        "Germany",
        "Germany",
        "Germany",
        "Germany",
        "Germany",
        "Germany",
        "France",
        "France",
    ],

    # Will be stripped
    "name": [
        " Alice ",
        " Bob",
        "Charlie ",
        " David ",
        " Eve",
        " Frank ",
        " Grace",
        " Heidi ",
        None,
        None,
    ],

    # Entirely empty column
    "empty_column": [None] * 10,
})

# Add a duplicate row
df.loc[10] = df.loc[0]

# Add a completely empty row
df.loc[11] = [None] * len(df.columns)
```

```python
from pprint import pprint
import danda  # noqa: F401  # registers the pandas accessor
result = df.dg.clean()
pprint(result.dg.report, sort_dicts=False)
```

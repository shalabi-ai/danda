```python
import numpy as np
import pandas as pd
# notice the need to danda import to have accessor df
import danda # noqa: F401  # registers the pandas accessor
```

```python
df = pd.DataFrame({
    "name": [
        "Alice",
        "   ",      # -> "" -> pd.NA
        "Bob",
    ],
    "city": [
        "Berlin",
        " NULL ",   # -> "NULL" -> pd.NA
        "London",
    ],
    "empty_column": [
        " ",
        " ",
        " ",
    ],
})
df
```

```python
from pprint import pprint

clean_df = df.dg.clean()
pprint(clean_df.dg.report)
clean_df
```

```python
df = pd.DataFrame(
    {
        "name": [
            " Alice ",    # leading/trailing whitespace
            "Bob",
            "Bob",        # duplicate row
            None,         # empty row (together with other None values)
            "Charlie",
        ],
        "city": [
            " London ",
            "Paris",
            "Paris",      # duplicate row
            None,
            " Berlin",
        ],
        "age": [
            30,
            25,
            25,           # duplicate row
            None,
            40,
        ],
        "empty_column": [
            np.nan,
            np.nan,
            np.nan,
            np.nan,
            np.nan,
        ],
    }
)

# Make row 3 completely empty
df.loc[3] = [None, None, None, None]

print(df)
```

```python
from pprint import pprint

clean_df = df.dg.clean()
pprint(clean_df.dg.report)
clean_df
```

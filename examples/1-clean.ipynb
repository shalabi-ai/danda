## clean
This example shows empty rows clean

```python
import numpy as np
import pandas as pd
import danda

data = pd.DataFrame({
    "Name": ["Alice", np.nan, "Charlie", np.nan, "Eve"],
    "Age": [25, np.nan, np.nan, np.nan, 30],
    "City": ["New York", np.nan, "Chicago", np.nan, np.nan],
})
data = data.dg.clean()
print(data.dg.report)
data
```

```python
# Empty columns
data = pd.DataFrame({
    "A": [1, 2, 3],
    "B": [np.nan, np.nan, np.nan],  # Entirely empty
    "C": [4, np.nan, 6],
    "D": [np.nan, np.nan, np.nan],  # Entirely empty
})
data = data.dg.clean()
print(data.dg.report)
data
```

```python
# drop duplicates
df = pd.DataFrame({
    "A": [1, 2, 2, 3],
    "B": ["a", "b", "b", "c"],
})

df = df.dg.clean()
print(df.dg.report)
df
```

```python
# clean white spaces in string columns
df = pd.DataFrame(
    {
        "name": [" Alice ", "Bob  ", "  Charlie"],
        "city": [" London", "Paris ", " Berlin "],
    }
)
df = df.dg.clean()
print(df.dg.report)
df
```

```python
import pandas as pd
import danda  # noqa: F401  # registers the pandas accessor
```

```python
df = pd.DataFrame({
    # Numeric
    "age": [20, None, 30, None, 40],

    # Boolean (converted by optimize())
    "approved": ["TRUE", None, "FALSE", "true", None],

    # Category (low cardinality)
    "department": ["IT", "HR", None, "IT", "HR"],

    # Text (high cardinality)
    "city": [
        "Berlin",
        None,
        "Paris",
        "London",
        "Madrid",
    ],

    # Datetime
    "hire_date": [
        "2024-01-01",
        None,
        "2024-01-03",
        None,
        "2024-01-05",
    ],
})
print(df.info())
df

```

```python
from pprint import pprint
df.dg.config.types.category_threshold =.5
result = (
    df
    .dg.clean()
    .dg.optimize()
)
pprint(result.dg.report)
print(result.info())
result
```

```python
result.dg.config.imputation.enabled = True
result1 = result.dg.impute()
result1
```
